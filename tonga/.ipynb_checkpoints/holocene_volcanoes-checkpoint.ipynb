## Filter, buffer or calculate distances using a while loop

### The structure of the following code is very simple:

```
    open the volcano dataset
        set the first volcano as the starting point for the loop
        **while volcano feature is not None**
            check if the current volcano is of interest
            **increment to new volcano**
```

````python
# open the input volcano points as a shapefile
driver = ogr.GetDriverByName("ESRI Shapefile")  # specify the type of file
input_data_source = driver.Open(volcanoes_shapefile, 0)  # open the file

counter = 0

if input_data_source is None:  # checks whether the file supplied exists and has read-in correctly
    print("Could not open the input shapefile.")
else:
    input_layer = input_data_source.GetLayer()  # get the input layer from the data source

    input_feature = input_layer.GetNextFeature()  # set the starting point as the first volcano

    while input_feature is not None:
        # get the volcano type
        volcano_type = input_feature.GetField("TYPE_")

        # check if the volcano type is in the specified list
        if volcano_type in ['Caldera', 'Calderas', 'Explosion', 'Stratovol']:
            counter += 1
            # space to do some calculations

        # this is the part that increments the while loop!
        # this will return **None** after incrementing over every volcano
        # it guarantees that we exit the loop
        input_feature = input_layer.GetNextFeature()  

    # close the input files
    input_data_source = None

print(f"\tfound {counter} volcanoes of interest of {volcanoes.shape[0]} total in dataset")
````

````python
# full code to buffer volcanoes of interest
from osgeo import ogr

# open the input shapefile
driver = ogr.GetDriverByName("ESRI Shapefile")
input_data_source = driver.Open(volcanoes_shapefile, 0)

counter = 0

if input_data_source is None:
    print("Could not open the input shapefile.")
else:
    # get the input layer
    input_layer = input_data_source.GetLayer()

    # open the output shapefile for writing
    output_shapefile_path = r"./volcanoes_of_interest_polygons.shp"
    output_data_source = driver.CreateDataSource(output_shapefile_path)
    
    # specify the output layer with polygon geometry type
    output_layer = output_data_source.CreateLayer("Volcanoes_Buffered", geom_type=ogr.wkbPolygon)

    # define fields for the output layer; this is a great section for a function!
    output_layer.CreateField(ogr.FieldDefn("NAME_", ogr.OFTString))
    output_layer.CreateField(ogr.FieldDefn("LOCATION", ogr.OFTString))
    output_layer.CreateField(ogr.FieldDefn("LAT", ogr.OFTReal))
    output_layer.CreateField(ogr.FieldDefn("LON", ogr.OFTReal))
    output_layer.CreateField(ogr.FieldDefn("ELEV", ogr.OFTReal))
    output_layer.CreateField(ogr.FieldDefn("TYPE_", ogr.OFTString))

    # initialize the feature variable
    input_feature = input_layer.GetNextFeature()

    while input_feature is not None:
        # get the volcano type
        volcano_type = input_feature.GetField("TYPE_")

        # check if the volcano type is in the specified list
        if volcano_type in ['Caldera', 'Calderas', 'Explosion', 'Stratovol']:
            counter += 1
            
            # get the point geometry
            point_geometry = input_feature.GetGeometryRef()

            # buffer the point by 1 kilometer
            buffered_geometry = point_geometry.Buffer(1)

            # create a new feature with the buffered geometry
            output_feature = ogr.Feature(output_layer.GetLayerDefn())
            output_feature.SetGeometry(buffered_geometry)

            # Set attribute values for the output feature
            output_feature.SetField("NAME_", input_feature.GetField("NAME_"))
            output_feature.SetField("LOCATION", input_feature.GetField("LOCATION"))
            output_feature.SetField("LAT", input_feature.GetField("LAT"))
            output_feature.SetField("LON", input_feature.GetField("LON"))
            output_feature.SetField("ELEV", input_feature.GetField("ELEV"))
            output_feature.SetField("TYPE_", input_feature.GetField("TYPE_"))

            # Add the feature to the output layer
            output_layer.CreateFeature(output_feature)

            # Clean up
            output_feature = None

        # move to the next feature in the input layer
        input_feature = input_layer.GetNextFeature()

    # Close the input and output shapefiles
    input_data_source = None
    output_data_source = None

print(f"buffered {counter = } points of {volcanoes.shape[0]} total in dataset")
````

````python
import math

# vancouver, bc coordinates
city_lat = 49.128
city_lon = -123.1207

def euclidean_distance(lat1, lon1, lat2, lon2):  
    """
        typical distance formula
    """
    return math.sqrt((lat1 - lat2) ** 2 + (lon1 - lon2) ** 2)

# pre-allocate variables to store the minimum distance and associated volcano name
min_distance = float('inf')  # set to positive infinity initially
min_distance_name = ""

# open the input shapefile
driver = ogr.GetDriverByName("ESRI Shapefile")
input_dataSource = driver.Open(volcanoes_shapefile, 0)

if input_dataSource is None:
    print("Could not open the input shapefile.")
else:
    # get the input layer
    input_layer = input_dataSource.GetLayer()

    # initialize the feature variable
    input_feature = input_layer.GetNextFeature()

    while input_feature is not None:
        # get the coordinates of the volcano point
        volcano_lat = input_feature.GetField("LAT")
        volcano_lon = input_feature.GetField("LON")
        volcano_name = input_feature.GetField("NAME_")

        # calculate the euclidean distance
        distance = euclidean_distance(volcano_lat, volcano_lon, city_lat, city_lon)

        # update the minimum distance and associated volcano name if a smaller distance is found
        if distance < min_distance:
            min_distance = distance
            min_distance_name = volcano_name

        # move to the next feature in the input layer
        input_feature = input_layer.GetNextFeature()

    # close the input shapefile
    input_dataSource = None

# report the minimum distance and associated volcano name
print(f"Minimum distance to Vancouver: {min_distance * 100:.2f} km")  # 1 degree latitude ~ 100 km
print(f"Volcano Name: {min_distance_name}")
````

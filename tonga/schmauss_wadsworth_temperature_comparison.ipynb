```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# comparing arrhenius temperature zones for Schmauss 2015 and Wadsworth (Delmelle) 2018/2021

# eruption temperature ranges
temperatures = np.linspace(-80, 800, 101)
kelvin = temperatures + 273.15

# temperatures
```

```python
gas_constant = 8.314e-3  # kJ/mol-K

# wadsworth says these should be less than 40 kJ/mol
adsorption_enthalpies = {'rhyolite': [-7.6 - 1., -7.6, -7.6 + 1.],  # kJ/mol
                         'dacite': [-17.8 - 3.1, -17.8, -17.8 + 3.1],
                         'andesite': [-13.7 - 3., -13.7, -13.7 + 3.]}

activation_energies = {'monodisperse': [133.79, 104.81, 80.]}

schmauss_arrhenius = [
    np.exp(-adsorption_enthalpies['rhyolite'][ii]/(gas_constant*kelvin)) 
    for ii in range(len(adsorption_enthalpies['rhyolite']))
]
wadsworth_arrhenius = [
    np.exp(-activation_energies['monodisperse'][ii]/(gas_constant*kelvin)) 
    for ii in range(len(activation_energies['monodisperse']))
]
```

```python
# plot the two on the same figure
fig, ax = plt.subplots(figsize=(11, 6))
ax2 = ax.twinx()

for ii, (schmauss, wadsworth) in enumerate(zip(schmauss_arrhenius, wadsworth_arrhenius)):
    ax.plot(temperatures, schmauss, 
            label=f"$\Delta$H = {adsorption_enthalpies['rhyolite'][ii]}")
    ax2.plot(temperatures, wadsworth, 
            label=f"$E_a$ = {activation_energies['monodisperse'][ii]}")
    
    
ax.set_title("SO$_2$ reaction rates for\nlow and high temperature mechanisms")
ax.set_xlabel("eruption temperatures ($^\circ$C)")
ax.legend(loc="upper left", title="Cold: physisorption")
ax2.legend(loc='upper right', title="Hot: chemisorption")
plt.tight_layout()
ax.set_ylabel("arrhenius reaction rates");
```
